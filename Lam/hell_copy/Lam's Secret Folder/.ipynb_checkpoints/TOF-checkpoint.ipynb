# Information from e-mail

The first exercise consists of a root file and a python script. Inside the root file you will find 8 ttrees. The ttrees contain emulated signals from a would-be experiment of two scintillators placed some distance apart and different particles flying through the two detectors, just as if the two detectors were in a beamline. In fact, what the ttree contains for each event is the time (relative to some reference) when a signal was seen in each of the two detectors. There are 4 different momenta and two different distances, one distance of 0 and another unknown distance. We would give you the challenge of taking these 8 ttrees and calculating the following things: what is the unknown distance, what is the fraction of protons to the other particles as a function of the momenta, can you calculate the fraction of protons for all momenta? The python script opens the attached ROOT file and prints out the contents of the file. The python script is called PrintFile.py and the ROOT file is called PseudoExperiment.root

```python
# Load TFile class from the ROOT library
from ROOT import TFile

# Load the root file
inputFile = TFile("PseudoExperiment.root", "READ");
```

```python
# Print some information on the root file
inputFile.Print()
```

```python
# list the contents of the root file
inputFile.ls()
```

You can see that in fact the TFile contains 8 TTrees.

In addition to a name (such as p500d0), each tree also has a description associated. In this case the description tells us more about the TTree

Let's try to find more information about these TTrees, we will start with the tree p500d0 and then move on to p500d_

```python
# First get the TTree from the TFile
p500d0 = inputFile.Get("p500d0")

# Then try to list it
p500d0.ls()
```

There is no addition information from the listing.

But we can print the tree contents, see below:

```python
# Print the ttree contents
p500d0.Print()
```

We now see much more information.

We have a table, where the first row, the header, contains information on the total size of the TTree in bytes as well as how many entries are in the TTree.

Subsequent rows in the table show the branches in the ttree. Each Branch has its own number of entries as well as a summary on the space consumption. The description of the branch (eg: Detector1/D) also tells us the type of data kept in the branch (see: https://root.cern.ch/doc/master/classTTree.html).

We know from the description of the exercise that these branches should contain information on the time when a particle traversed each detector. Lets try to make a plot of the arrival times at detector 1

```python
# In order to draw, we need to draw on a canvas
# import the canvas from root and create it
from ROOT import TCanvas
c1 = TCanvas("c1", "c1", 800, 600)

# Plot an anonymous histogram
p500d0.Draw("Detector1")
c1.Draw()
```

This plot is a histogram.
* Does it look good?
* What if we wanted to plot the second detector?
* What about putting labels on the axis?

For this, we need to manipulate the histogram itself.

```python
# Import the histogram class
from ROOT import TH1D
from ROOT import kRed

hist1 = TH1D("hist1", "Detector 1;time ns;Events", 50, 35, 85)
hist2 = TH1D("hist2", "Detector 2;time ns;Events", 50, 35, 85)
p500d0.Draw("Detector1>>hist1")
p500d0.Draw("Detector2>>hist2")

# Create a new canvas for this histogram
c2 = TCanvas("c2", "c2", 800, 600)
hist1.Draw()
hist2.SetLineColor(kRed)
hist2.Draw("same")

c2.Draw()
```

Lets start with calculating the unknown distance

```python

tofHist = TH1D("tofHist", "Time of flight", 50, -35, 35)
p500d0.Draw("Detector2 - Detector1>>tofHist")

# Create a new canvas for this histogram
c3 = TCanvas("c3", "c3", 800, 600)
tofHist.Draw()

c3.Draw()
```

It seems there is a time offset of -8 ns. If we want to accurately calculate the distance, we need to correct for this.

Lets take a look with the unknown distance:

```python
# Get the TTree for the unknown distance
p500d_ = inputFile.Get("p500d_")

# Plot the TOF
c4 = TCanvas("c4", "c4", 1000, 500)
c4.Divide(2,1)
c4.cd(1)
p500d_.Draw("Detector1")
c4.cd(2)
tofHistUn = TH1D("tofHistUn", "Time of flight", 50, 20, 110)
p500d_.Draw("Detector2 - Detector1 + 8.004>>tofHistUn")
tofHistUn.Draw()

c4.Draw()
```

There is a second peak... what is it?

```python
# Remove the peak
tofHistUnFix = TH1D("tofHistUnFix", "Time of flight", 50, 20, 110)
p500d_.Draw("Detector2 - Detector1 + 8.004>>tofHistUnFix", "Detector2 - Detector1 + 8.004 < 70")

# Get Mean
mean = tofHistUnFix.GetMean()
print(mean, "ns")

# Calculate distance, electrons at this speed are effectively at the speed of light
c = 3 * 10**8
distance = c * mean * 10**(-9)
print("The distance is:", distance, "m")

# We can also use the drawing to do our calculation:
c5 = TCanvas("c1", "c1", 1000, 400)
c5.Divide(3,1)

c5.cd(1)
tofHistUnFix.Draw()

c5.cd(2)
p500d_.Draw("(Detector2 - Detector1 + 8.004) * 3 / 10", "Detector2 - Detector1 + 8.004 < 70")

c5.cd(3)
p500d_.Draw("(Detector2 - Detector1) * 3 / 10", "Detector2 - Detector1 + 8.004 < 70")

c5.Draw()
```
