```python
#Import necessary libraries
from ROOT import TCanvas, TH2D
import numpy as np
import matplotlib.pyplot as plt

c1=TCanvas("MimosaPlots", "MimosaPlots", 1600, 600)
c1.Divide(2,1)

c1.cd(1)
hist1=TH2D("name", "Raw Data", 1152, 0, 1152, 576, 0, 576)
hist2=TH2D("anothername", "Data minus Noisy?", 500, 0, 1152, 250, 0, 576)

RAWdata.Draw("Mimosa0_hitY:Mimosa0_hitX>>name") #Draw all the raw data from Mimosa0 to a histogram

hitCountsTot=[] #create an array to record the number of hits each pixel records
for i in range(0, 1152): #hist1.GetNbinsX()
    for j in range(0, 576): #hist1.GetNbinsY()
        hitCountsTot+=[hist1.GetBinContent(i, j)] #Iterate through all bins (where # of bins is equal to the # of pixels)
                                                    #And add to the list of pixel hit numbers

meanHitCount=np.mean(hitCountsTot) #get the mean pixel count
noisyPixels=set() #make a set to contain noisy pixel locations
hitCounts=[] #create an array to record the number of hits each noisy pixel records
for i in range(0, 1152): 
    for j in range(0, 576): 
        if hist1.GetBinContent(i, j) > meanHitCount*20: #if the number of hits is 20x the mean number of pixel hits
            noisyPixels.add((i, j)) #add the pixel to our list of noisy pixels
            hitCounts+=[hist1.GetBinContent(i, j)] #and add its number of hits to our array
print(noisyPixels, len(noisyPixels)) #print the set of noisy pixels
print("Mean is", np.mean(hitCounts), "with range", np.ptp(hitCounts), "from", np.amin(hitCounts), "to", np.amax(hitCounts)) #Print information about the noisy pixels

pad2=c1.cd(2) #canvas.cd(#) returns a pad
pad2.SetLogz() #this helps by setting the colz coloring to a log scale
currentPixel=()
for event in RAWdata:
    for j in range(0, event.NMimosa0_hitX): #as the array lengths for x and y should be the same
        currentPixel=(event.Mimosa0_hitX[j], event.Mimosa0_hitY[j])
        if currentPixel not in noisyPixels:
            hist2.Fill(currentPixel[0], currentPixel[1]) #adds that data point to our histogram

hist2.Draw("colz") #color it!

c1.Draw()
```

```python
#Now we want to look at noisy pixels across multiple runs to see if they are the same ones
runNumbers=["1571933143", "1571931737", "1571932015", "1571932325", "1571932668", "1571933143", "1571933521", "1571933929", "1571934315", "1571934931", "1571935652", "1571936276", "1571936885", "1571937854", "1571938197", "1571938501", "1571938794", "1571939126", "1571939524", "1571939912", "1571950010", "1571950464", "1571950975", "1571951573", "1571952203", "1571952846", "1571953473"]

noisyPixelsDict={}
c1=TCanvas("MimosaPlots", "MimosaPlots", 1600, 600)

for run in runNumbers:
#for i in range(0,1):
    runName="/nfs/dust/fhlabs/group/BL4S/data/DESYChain/ConvertedData/" + run + ".root"
    importFile = TFile(runName, "READ")
    RAWdata=importFile.Get("RAWdata")
    hist1=TH2D("name", "Raw Data", 1152, 0, 1152, 576, 0, 576)
    RAWdata.Draw("Mimosa0_hitY:Mimosa0_hitX>>name") #Draw all the raw data from Mimosa0 to a histogram
    
    hitCountsTot=[]
    for i in range(0, 1152): 
        for j in range(0, 576): 
            hitCountsTot+=[hist1.GetBinContent(i, j)]
    
    meanHitCount=np.mean(hitCountsTot) #get the mean pixel count
    noisyPixels=set() #make a set to contain noisy pixel locations
    hitCounts=[] #create an array to record the number of hits each noisy pixel records
    for i in range(0, 1152): 
        for j in range(0, 576): 
            if hist1.GetBinContent(i, j) > meanHitCount*20: #if the number of hits is 20x the mean number of pixel hits
                noisyPixels.add((i, j)) #add the pixel to our list of noisy pixels
                hitCounts+=[hist1.GetBinContent(i, j)] #and add its number of hits to our array
    noisyPixelsDict[run]=noisyPixels
    
```

```python
for i in range(0, len(runNumbers)-1): #show the pixels that are in one run but not the consecutive one
    print(len(noisyPixelsDict[runNumbers[i+1]]))
    print(len(noisyPixelsDict[runNumbers[i]]))
    print(noisyPixelsDict[runNumbers[i+1]]-noisyPixelsDict[runNumbers[i]])
    print()
```

```python
#Make a dictionary referenced by pixel (the key is the pixel tuple) saying in how many runs it was
import math 
pixelNums=dict()
for run in runNumbers:
    for pixelSet in noisyPixelsDict[run]: 
        try:
            pixelNums[pixelSet]+=1
        except:
            pixelNums[pixelSet]=0
print(len(runNumbers))
print(np.mean([i for i in pixelNums.values()]))
```

```python
#Now see how many pixels are in most or all of the runs
necNumber=math.floor(len(runNumbers)*0.7)
mostlyPixels=0 #those that are in at least 70% of the runs
alwaysPixels=0 #those that are in all the runs
sometimesPixels=0 #those that are in 30% or less of the runs

mostlyPix=set()

for pixel in pixelNums:
    if pixelNums[pixel]>=necNumber:
        #print(pixelNums[pixel])
        mostlyPix.add(pixel)
        mostlyPixels+=1
        if pixel[1]>=len(runNumbers)-1: #in this case one of the runs seemed to have very few noisy pixels, so I'm effectively dis-including it 
            alwaysPixels+=1
    else:
        sometimesPixels+=1
```

```python
#And prin the percentages of pixels
print(mostlyPixels/(mostlyPixels+sometimesPixels)*100)
print(alwaysPixels/(mostlyPixels+sometimesPixels)*100)
print(alwaysPixels/(mostlyPixels)*100)
```

```python
#Plus graph the mostly noisy pixels, for good measure
plt.scatter([i[0] for i in mostlyPix], [i[1] for i in mostlyPix]) #show where the noisy pixels
plt.xlabel('x') 
plt.ylabel('y')
```
